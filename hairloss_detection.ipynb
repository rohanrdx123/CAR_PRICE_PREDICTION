```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np

originalImage = cv2.imread('/1.jpg')
grayImage = cv2.cvtColor(originalImage, cv2.COLOR_BGR2GRAY)
  
(thresh, blackAndWhiteImage) = cv2.threshold(grayImage, 127, 255, cv2.THRESH_BINARY)
cv2.imwrite('result.jpg',blackAndWhiteImage)
cv2_imshow(blackAndWhiteImage)

img=cv2.imread('/content/result.jpg')
white_pix=np.sum(img==255)
black_pix=np.sum(img==0)
print("WHITE: ",white_pix)
print("BLACK: ",black_pix)
sum= white_pix + black_pix

hairloss=white_pix/sum

percent=hairloss*100

print("hairloss percentage: "+str(percent)+"%")
```
